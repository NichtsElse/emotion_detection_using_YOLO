```python
import os
import cv2
from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator, colors

# Define input and output video paths
input_video_path = '/content/02-01-08-02-02-02-01.mp4'
output_video_path = 'kaget.mp4'
model = YOLO("/content/best12.pt")

# Check if file exists
if not os.path.isfile(input_video_path):
    raise FileNotFoundError(f"Input video file not found at: {input_video_path}")

# Open the input video
cap = cv2.VideoCapture(input_video_path)

# Check if video capture opened successfully
if not cap.isOpened():
    raise ValueError(f"Error opening video file: {input_video_path}")

# Get the video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Use 'mp4v' for MP4 format
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Define emotion labels
emotion_labels = {
    0: "afraid",
    1: "anger",
    2: "disgusting",
    3: "happy",
    4: "neutral",
    5: "sad",
    6: "surpise"
}

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Perform inference
    results = model(frame)

    # Iterate through results
    # results[0] is now a Result object
    annotated_frame = results[0].plot()

    # Access boxes from the Result object using .boxes attribute
    for box in results[0].boxes:
        # Get bounding box coordinates
        xyxy = box.xyxy[0].tolist()  # Convert tensor to list
        conf = box.conf[0].item()  # Get confidence score
        cls = box.cls[0].item()  # Get class index

        # Get emotion label
        emotion_label = emotion_labels.get(int(cls), "unknown")

        # Add emotion label to annotated frame
        label = f'{emotion_label} {conf:.2f}'

        # Use Annotator to draw bounding boxes
        annotator = Annotator(annotated_frame)
        annotator.box_label(xyxy, label, color=colors(int(cls), True))

    # Write the frame to the output video
    out.write(annotated_frame)

# Release everything
cap.release()
out.release()

print(f"Processed video saved to {output_video_path}")
```
